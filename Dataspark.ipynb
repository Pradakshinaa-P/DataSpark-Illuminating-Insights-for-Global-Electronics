```python
import pandas as pd

import pandas as pd
def readdata():
  try:
    x=input("enter the file name")
    df=pd.read_csv(x)
    return df
  except:
    df=pd.read_csv(x,encoding='latin-1')
    return df

#func to find missing values
def findnull():
  x=input("enter the file name")
  df=pd.read_csv(x)
  return df.isnull().sum()


#func to find duplicate values
def finddup():
  x=input("enter the file name")
  df=pd.read_csv(x)
  return df.duplicated().sum()
  

#func to find datatypes
def findtype():
  x=input("enter the file name")
  df=pd.read_csv(x)
  return df.dtypes


#func to load the cleanes dataset
def loadcsv(str):
  x=input("enter the file name")
  y=str.to_csv(x,index=False)
  return y 

```
